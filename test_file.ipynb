# GOAL: Add a team-level target variable from team_stats to player_stats
#       and analyze which individual player stats affect it most
#
# STEPS:
# 1. Data Preparation - Verify team name matching between datasets
# 2. Merge Function - Create function to merge team_stats column to player_stats
# 3. Target Variable Selection - Display available team_stats columns
# 4. Correlation Analysis - Calculate correlations between player stats and target
# 5. Feature Importance - Rank features by impact on target variable
# 6. Visualization - Create plots showing top features affecting target
# 7. Reusable Function - Quick analysis function for any target variable


````python
# Step 2: Function to add target variable from team_stats to player_stats
def add_target_variable(player_df, team_df, target_column, player_team_col='Team', team_name_col='Team Name'):
    """
    Merge a target variable from team_stats to player_stats
    
    Parameters:
    - player_df: player_stats dataframe
    - team_df: team_stats dataframe  
    - target_column: name of column in team_df to add as target
    - player_team_col: column name in player_df with team names (default: 'Team')
    - team_name_col: column name in team_df with team names (default: 'Team Name')
    
    Returns:
    - player_df with target variable added
    """
    # Select only Team Name and target column from team_df
    team_subset = team_df[[team_name_col, target_column]].copy()
    
    # Merge
    merged_df = player_df.merge(
        team_subset, 
        left_on=player_team_col, 
        right_on=team_name_col, 
        how='left'
    )
    
    # Rename target column to include '_target' suffix for clarity
    target_name = f'{target_column}_target'
    merged_df = merged_df.rename(columns={target_column: target_name})
    
    # Drop the team name column from team_df (we already have it in player_df)
    merged_df = merged_df.drop(columns=[team_name_col])
    
    print(f"✓ Added '{target_name}' to player_stats")
    print(f"  Rows: {len(merged_df)}")
    print(f"  Missing values: {merged_df[target_name].isna().sum()}")
    
    return merged_df

# Example usage (commented out):
# player_stats = add_target_variable(player_stats, team_stats, 'Goal Difference')
````
